### 2. Removing Unnecessary Columns and Renaming Columns for Clarity

```python
# Removing unnecessary columns

df=df.drop(columns = ['birthDate', 'date', 'status', 'tax_revenue_country_country', 'source', 'category', 
                      'lastName', 'firstName', 'birthYear', 'birthMonth', 'birthDay'])
```

```python
# Checking that 12 columns were dropped

df.shape
```

```python
# Investigating Missing Values

df.isnull().sum()
```

### There are too many missing values in the 'organization' and 'title' columns to be useful for analysis. So, these columns will be removed. 

### The 'state' and 'residenceStateRegion' are missing many values, however, the values it does contain apply to all billionaires in the United States and may be useful for analyzing that region. So, those will be left as they are. 

### The next largest number of values missing for a field is 184. This is less than 10% of the 2640 records and those fields can still be useful for analysis even with those missing values. So those will be left as they are.

```python
# Removing columns with mostly missing values

df=df.drop(columns = ['organization', 'title'])
```

```python
# Checking that 2 more columns were dropped

df.shape
```

```python
# Renaming columns for clarity

df.rename(columns = {'finalWorth' : 'final_worth', 'personName' : 'name', 'countryOfCitizenship' : 'citizenship_country', 
                    'selfMade' : 'self_made', 'residenceStateRegion' : 'state_region', 
                     'gross_tertiary_education_enrollment' : 'tertiary_ed_enroll', 
                     'gross_primary_education_enrollment_country' : 'primary_ed_enroll'}, inplace = True)
```

### 3. Looking for mixed data types and fixing data types

```python
# Looking for mixed data types

for col in df.columns.tolist():
  weird = (df[[col]].applymap(type) != df[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (df[weird]) > 0:
    print (col)
```

### These columns with mixed data types are all strings with some data missing. I will not change the data types on these columns to make them completely strings because when doing that, the missing values no longer appear as missing when running this script: df.isnull().sum(). 
### The data that is there is the correct data type, with the exception of 'gdp_country'. That field needs to be converted to int64 or float64.

```python
# Removing dollar signs from gdp_country column (received next line of code from ChatGPT)

df['gdp_country'] = df['gdp_country'].replace('[\$,]', '', regex=True).astype(float)
```

```python
# Checking that dollar signs were removed and data type changed.

df['gdp_country'].head()
```
